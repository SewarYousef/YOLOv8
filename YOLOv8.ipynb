# YOLOv8 Tasks

YOLOv8 can train, val, predict and export models for the most common tasks in vision AI: Classify, Detect, Segment, and Pose

banner-tasks.png

##1. Classify

YOLOv8 classification models use the -cls suffix, such as yolov8n-cls.pt and are pretrained on ImageNet

```python
%cd {HOME}
!yolo task=classify mode=predict model=yolov8n-cls.pt conf=0.25 source='https://ultralytics.com/images/bus.jpg' save=True

# conf=0.25 Sets the confidence threshold for detections to %25
# Which means only detections with confidence scores higher than 0.25 will be considered
```

```python
%cd {HOME}
Image(filename='runs/classify/predict/bus.jpg', height=600)
```

##2. Detect

YOLOv8 detection models have no suffix and are the default YOLOv8 models, such as `yolov8n.pt` and are pretrained on COCO.


```python
%cd {HOME}
!yolo task=detect mode=predict model=yolov8n.pt conf=0.25 source='https://ultralytics.com/images/bus.jpg' save=True
```

```python
%cd {HOME}
Image(filename='runs/detect/predict/bus.jpg', height=600)
```

##3. Segment

YOLOv8 segmentation models use the -seg suffix, such as yolov8n-seg.pt and are pretrained on COCO

```python
%cd {HOME}
!yolo task=segment mode=predict model=yolov8n-seg.pt conf=0.25 source='https://ultralytics.com/images/bus.jpg' save=True
```

```python
%cd {HOME}
Image(filename='runs/segment/predict/bus.jpg', height=600)
```

##4. Pose
